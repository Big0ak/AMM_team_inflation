### Предсказывать теперь надо только цены этих продуктов

```python
def Creat_line_regression(df_full, model_list, error_test):
    for element in df_full:
        data = df_full[element].copy()
        data.drop(['new_WebPriceId'], axis=1, inplace=True)
        
        make_features(data, 6, 6) 
        train, test = train_test_split(data, shuffle = False, test_size = 0.2)
        train = train.dropna() # удаляет пустые значения
        
        X_train = train.drop('new_CurrentPrice', axis = 1)
        y_train = train['new_CurrentPrice']
        X_test = test.drop('new_CurrentPrice', axis = 1)
        y_test = test['new_CurrentPrice']
        
        model = LinearRegression().fit(X = X_train, y = y_train)
        error_test.append(mean_absolute_error(y_test, model.predict(X_test)))
        model_list.append(model)
```

```python
def make_features(data, max_lag, rolling_mean_size):
    data['year'] = data.index.year
    data['month'] = data.index.month
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['new_CurrentPrice'].shift(lag)
    
    data['rolling_mean'] = data['new_CurrentPrice'].rolling(rolling_mean_size).mean()
```
